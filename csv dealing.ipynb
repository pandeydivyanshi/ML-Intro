## 3. opening a csv file from url

```python
import requests
import pandas as pd
from io import StringIO
url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv"
headers = {
    "User-Agent": "Mozilla/5.0"
}

req = requests.get(url, headers=headers)
data = StringIO(req.text)

df = pd.read_csv(data)
df.head()
```

## 14.convertors

```python
import pandas as pd

def rename(name):
    if name == "credit":
        return "C"
    else:
        return name

df = pd.read_csv(
    'customer_churn_dataset.csv',
    converters={'payment_method': rename}
)
df
```

## 15. na_value parameter

na_values specifies which values in the dataset should be recognized as missing data and converted to NaN during import.

```python
pd.read_csv(
    'customer_churn_dataset.csv',
    na_values=['NA', 'N/A', 'null', ' ', '']
)
```
