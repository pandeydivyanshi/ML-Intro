## API

API (Application Programming Interface) is a way for one software to request data or services from another system in a structured format.

Your Code (Request)
        ↓
     API Server
        ↓
   JSON Response
        
Python → pandas DataFrame
API is like datapipeline

```python
import requests

url = "https://api.themoviedb.org/3/movie/changes"
params = {
    "api_key": "8a2327bfb60bcf237edfab8e1436d1b7",
    "page": 1
}

response = requests.get(url, params=params)
print(response.status_code)
print(response.json())

```

```python
import pandas as pd
pd.DataFrame(response.json()['results'])
```
