## 2. import json from url

```python
import requests
import pandas as pd

# Public API that returns JSON
url = "https://jsonplaceholder.typicode.com/users"

response = requests.get(url)

# Check response
print(response.status_code)   # should be 200

# Convert JSON to Python object
data = response.json()

# Convert to DataFrame
df = pd.json_normalize(data)

df.head()
```

## working with SQL 

MySQL (XAMPP)
     ↓
pandas.read_sql()
     ↓
Data Cleaning & EDA
     ↓
Feature Engineering
     ↓
ML Model


```python
# STEP 1: Import required libraries
import pandas as pd
from sqlalchemy import create_engine

# STEP 2: Create MySQL connection (XAMPP default settings)
# username: root
# password: empty
# host: localhost
# port: 3306
# database: replace with your database name

engine = create_engine(
    "mysql+pymysql://root:@localhost:3306/your_database_name"
)

# STEP 3: Read SQL table into pandas DataFrame
df = pd.read_sql("SELECT * FROM your_table_name", engine)

# View first few rows
print(df.head())

# STEP 4: Run a custom SQL query
query = """
SELECT *
FROM your_table_name
LIMIT 10
"""
df_query = pd.read_sql(query, engine)
print(df_query)

# STEP 5: Write DataFrame back to MySQL (optional)
df.to_sql(
    name="new_table_name",
    con=engine,
    if_exists="replace",   # use "append" to add rows
    index=False
)

print("SQL connection and data transfer successful!")
```
