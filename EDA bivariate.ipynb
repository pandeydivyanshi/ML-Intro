### 4.Distplot  (NUmerical-categorical)

```python
sb.distplot(tit['Age'])
```

```python
sb.kdeplot(tit[tit['Survived'] == 0]['Age'], label='Not Survived')
sb.kdeplot(tit[tit['Survived'] == 1]['Age'], label='Survived')
```
